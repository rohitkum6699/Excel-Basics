# Q1) Explain the difference between Relative , Absolute and Mixed Cell Referencing.
 You have the following dataset:
 Vehicle
 Car 500000
 Bicycle
 Price
 10500
 Auto-Rickshaw
 25000

 Ans-

To answer this question i have used given dataset in this question
to **demonstrate Relative, Absolute, and Mixed cell referencing in Excel**.  

---

## Dataset-
| **A**           | **B**       |
|-----------------|-------------|
| Vehicle         | Price       |
| Car             | 500000      |
| Bicycle         | 10500       |
| Auto-Rickshaw   | 25000       |

---


### 1. **Relative Referencing**
- **Notation:** A1
- **Behavior:** Adjusts automatically when copied to another cell.
- **Example:**
  - In cell C2, type =B2
  - Copy down to C3  becomes =B3 (Bicycleâ€™s price).
  - Copy down to C4  becomes =B4 (Auto-Rickshawâ€™s price).
- **Use case:** Applying the same formula across rows/columns.

---

### 2. **Absolute Referencing**
- **Notation:** `$A$1`
- **Behavior:** Always points to the same cell, no matter where copied.
- **Example:**
  - Suppose Carâ€™s price B2 is the base.
  - In C3, type =B3/`$B$2`
  - Copy down to C4 â†’ still =B4/`$B$2`
  - Result: Bicycle and Auto-Rickshaw prices compared to Carâ€™s price.
- **Use case:** Fixed references like tax rates, constants, or base values.

---

### 3. **Mixed Referencing**
- **Notation:** `$A1 or A$1`
- **Behavior:** Locks either row or column, while the other adjusts.
- **Examples:**
  - `$B2` â†’ Column **B** fixed, row changes when copied down.
  - `B$2` â†’ Row **2** fixed, column changes when copied across.
- **Use case:** Useful in tables where one dimension (row or column) must remain constant.

---


ðŸ‘‰ **Summary:**
- **Relative** â†’ moves with the formula.  
- **Absolute** â†’ locked in place.  
- **Mixed** â†’ half locked, half flexible.
